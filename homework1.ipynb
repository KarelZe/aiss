```python

# create selection as specified above
selection = ['Age', 'Est_Income', 'Local']

# apply standarization 
standard_scaler = StandardScaler()
df[selection] = standard_scaler.fit_transform(df[selection])

# print df info
df.describe()
```

```python
columns_org = df.columns
columns_pca=['PC1','PC2', 'PC3', 'PC4']

# standardize data
na = standard_scaler.fit_transform(df)
df = pd.DataFrame(na,columns=columns_org)

# use pca with 4 factors
pca = PCA(n_components=4)
df[columns_pca] = pca.fit_transform(df)

# calculate abs value to treat negative correlations as positive ones.
corr = df.corr().abs()
corr_sliced = corr.loc[columns_pca,columns_org]

# get column name of feature with max correlation for each pc
corr_sliced["max_corr_col"] = corr_sliced.idxmax(axis=1)

# get corresponding correlation to column
corr_sliced["max_corr_val"] =  corr_sliced.apply(lambda x: x[x["max_corr_col"]], axis=1)

# return only relevant columns
corr_sliced[["max_corr_col", 'max_corr_val']]
```
